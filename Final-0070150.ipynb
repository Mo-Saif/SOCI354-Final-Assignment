14- Extract all words from the 'text' field and put them in a new column. The findall method and a regular expression to recognize words will help you.

```python
df['words'] = df.text.str.findall(r'[\w]+')
```

15- Identify the words that occur the most. The Counter object from collections will help you.

```python
from collections import Counter
import itertools
```

```python
# list of all words in all tweets
words = list(itertools.chain.from_iterable(df['words']))
```

```python
c = Counter(words)
```

```python
print('Words that occur the most:')
c.most_common(30)
```

16- How many of the words occur only once? List 100 of them. Provide tweets that contain these words.

```python
oneoccs = 0
for i in c.items():
    if i[1] == 1:
        oneoccs += 1 
print('Number of words that occur only once: {}'.format(oneoccs))
```

```python
print('100 words that appeared only once: ')
c.most_common()[:-100-1:-1]
```

```python
# The 100 words without their number of occurences
words1 = [i[0] for i in c.most_common()[:-100-1:-1]]
```

```python
# Tweets that contain any of the 100 words
df[[any(i in x for i in words1) for x in df['words']]]
```

17- Find the longest 100 words. List them.

```python
import numpy as np
```

```python
unique_words = list(set(words))
```

```python
# Lengths of all words
lens = [len(i) for i in unique_words]
```

```python
# Indices of sorted words by length
inds = np.argsort(lens)
```

```python
print('List of longest 100 words: ')
[unique_words[i] for i in inds[-100:]]
```
